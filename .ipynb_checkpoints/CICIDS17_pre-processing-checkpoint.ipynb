```python
# Remove Problematic Values and Save to Parquet
import gc
import time

fbmean = 1491719.0643420687
fbmax = 2 * 2071000000.0
fpmean = 70854.23306262735
fpmax = 2 * 4000000.0


original_shape = df.shape
print(f"Loaded dataframe with shape: {original_shape}")
#print(df.columns)

# remove repeated header rows
repeated_header_rows = []
for col in df.columns:
    mask = df[col] == col
    repeated_header_rows.extend(df.index[mask].tolist())
    df = df[~mask]
repeated_header_rows = sorted(set(repeated_header_rows))
print(f"Removed {len(repeated_header_rows)} repeated header rows at indices: {repeated_header_rows}")

# Convert relevant columns to numeric
for col in ['Flow Bytes/s', 'Flow Packets/s']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# replace NaN and infinity values
if 'Flow Bytes/s' in df.columns:
    df['Flow Bytes/s'] = df['Flow Bytes/s'].replace(np.nan, fbmean)
    df['Flow Bytes/s'] = df['Flow Bytes/s'].replace(np.inf, fbmax)
if 'Flow Packets/s' in df.columns:
    df['Flow Packets/s'] = df['Flow Packets/s'].replace(np.nan, fpmean)
    df['Flow Packets/s'] = df['Flow Packets/s'].replace(np.inf, fpmax)

# Report columns and rows with NaN values
cols_with_nan = df.columns[df.isna().any()].tolist()
rows_with_nan = df.index[df.isna().any(axis=1)].tolist()
print(f"Columns with NaN values: {cols_with_nan}")
print(f"Rows with NaN values: {rows_with_nan[:10]}{'...' if len(rows_with_nan) > 10 else ''} (total: {len(rows_with_nan)})")

# Downcast datatypes for efficiency
for column_name in df.columns:
    if column_name in dtypes:
        dtype = dtypes[column_name]
        try:
            df[column_name] = df[column_name].astype(dtype)

        except Exception as col_err:
            print(f"Warning: Could not convert column '{column_name}' to {dtype} in file dataframe: {col_err}")

print(df.info())
# save as Parquet
df.to_parquet('datasets/cicids17/cicids17-pre-processed.parquet')
print(f"Saved cleaned file with shape {df.shape} as: datasets/cicids17/cicids17-pre-processed.parquet")

# Clean up
del df
gc.collect()

```
